```python
import pandas as pd

# ouvrir le fichier data\EDdA_dataframe_withContent.tsv
df = pd.read_csv('data/EDdA_dataframe_withContent.tsv', sep='\t')

# Supprimer les trois premières lignes
df = df.drop([0, 1, 2])
df
```

```python
# Quelles valeurs sont présentes dans la colonne 'author' ?
df['author'].unique()
```

```python
# combien de valeurs différentes sont présentes dans la colonne 'author' ?
df['author'].nunique()
```

```python
import re
df['author'] = df['author'].str.replace(r'(\w+)(\d+)$', r'\1', regex=True)

# Vérifiez les valeurs uniques après le nettoyage
print(df['author'].nunique())
```

```python
# Séparer les auteurs multiples
df['author'] = df['author'].str.split('&')
df = df.explode('author')
df
```

```python
# Choisir 500 articles au hasard
df = df.sample(50)

# Créer un fichier tsv avec les données
df.to_csv('data/EDdA_50dataframe.tsv', sep='\t', index=False)
df
```
